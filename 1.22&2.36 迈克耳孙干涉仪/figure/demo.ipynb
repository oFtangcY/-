```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 3, 4, 5, 6, 7, 8, 10, 13, 14])
y = np.array([49.93664, 50.09529, 50.176, 50.25595, 50.33496, 50.41395, 50.49527, 50.65645, 50.89157, 50.96922])

p = np.polyfit(x, y, deg=1)
a1=p[0]
a2=p[1]
y1=[]
for i in x:
    y1.append(a1*i+a2)

plt.scatter(x, y, marker='o', c = '#FF4500')
plt.plot(x, y1, '#FF8C00')

plt.show()
```
